## Checking the Number of outliers present in each features and Non outliers in each features
## Performing the IOR Method

```python

Feature_names = Boston_data.columns

Data = Boston_data

for features_names in Feature_names:
    print("Column name = " ,  features_names)
    data = Boston_data[features_names].sort_values()
    q25, q75 = np.percentile(data, 25), np.percentile(data, 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    
    outliers = [x for x in Data[features_names] if x < lower or x > upper]
    print('Identified outliers: %d' % len(outliers))
    
    
    outliers_removed = [x for x in Data[features_names] if x >= lower and x <= upper]
    print('Non-outlier observations: %d' % len(outliers_removed),"\n")

```

```python
INdependent_Data = Boston_data.drop("House_Price",axis=1)
Dependent_data = Boston_data["House_Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test = train_test_split(INdependent_Data , Dependent_data , test_size=0.20,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_GRID = {'copy_X': [True,False],
 'fit_intercept': [True,False],
 'n_jobs': [1,2],
 'normalize': [True,False],
 'positive': [True,False]}
```

```python
Grid_linear = GridSearchCV(estimator=LinearRegression(),verbose=5,param_grid=param_GRID)

Grid_linear
```

```python
Grid_linear.fit(X=X_train,y=y_train)
```

## Training The model

```python
Regressor = LinearRegression(normalize=True,n_jobs=1)

Regressor.fit(X=X_train,y=y_train,)
```

## Test Accuracy

```python
Regressor.score(X_test,y_test)
```

## Prediction

```python
Regressor.predict(X_test)
```
